```python
import numpy as np
import PIL
import random
from PIL import Image
from PIL import ImageDraw
```

```python
## IL.ImageDraw.ImageDraw.ellipse(xy, fill=None, outline=None, width=0)
image = Image.new("RGBA", (1000,1000))
draw = ImageDraw.Draw(image)
```

```python
x1 = 0
x2 = image.size[0]
y1 = 0
y2 = image.size[1]
fill_color = (85,85,85)
#outline_color = (10,10,70)
outline_color = None
length = 2000
fake_sin = [5,15,20,15,5,-5,-15,-20,-15,-5]
for i in range(length):
    width = 1
    draw.ellipse([(x1,y1), x2,y2], fill = fill_color, outline = outline_color, width = width)
    x1 = x1 +random.randint(0,2)
    x2 = x2 -random.randint(0,2)
    y1 = y1 +random.randint(0,2)
    y2 = y2 -random.randint(0,2)
    fill_color = (int(fill_color[0] +fake_sin[i%len(fake_sin)]) , int(fill_color[1] +fake_sin[i%len(fake_sin)]), int(fill_color[2]+fake_sin[i%len(fake_sin)]))
image
```
