```python
from PIL import Image
from PIL import ImageDraw
import cmath
import numpy as np
```

```python
image = Image.new("RGB", (800,800))
draw = ImageDraw.Draw(image)
xa = -1
xb = 1
ya= -1
yb = 1
for x in range(image.width):
    for y in range(image.height):
        x0 = image.width/(xa-xb)
        y0 = image.height/(xa-xb)
        z = complex(x0,y0)
        f = z*z
        
        R = int(f.real%255)
        G = int(f.imag%255)
        B = 0
        
        color = (R,G,B)
        image.putpixel((x,y), color)
```

```python
#Setup the image with appropriate proportions
#Setup drawing area
xa = 0
xb = 11
ya = 8
yb = 11

#setup image size with black background
imgX = 800
imgY = 200
image = Image.new("RGB", (imgX,imgY), (0,0,0))

for Py in range(imgY):
    y0 = Py * (yb-ya) / (imgY-1) +ya
    for Px in range(imgX):
        x0 = Px * (xb-xa) / (imgX-1) +xa
        z = complex(x0,y0)
        
        f = z*z
        
        R = int((f.real**2)%255)
        #R = 50
        G = int(f.imag%255)
        #G = 50
        B = 100
        
        color = (R,G,B)
        image.putpixel((Px,Py), color)
image
```

```python
#Setup the image with appropriate proportions
#Setup drawing area
xa = 25
xb = 35
ya = 25
yb = 35

#setup image size with black background
imgX = 500
imgY = 500
image = Image.new("HSV", (imgX,imgY), (0,0,0))

for Py in range(imgY):
    y0 = Py * (yb-ya) / (imgY-1) +ya
    for Px in range(imgX):
        x0 = Px * (xb-xa) / (imgX-1) +xa
        z = complex(x0,y0)
        
        f = z*z
        
        H = int(f.real)
        #R = 50
        #G = int(50*f.real%255)
        S = int(100*np.sin(5*f.imag)+150)
        V = int(50*np.cos(2*f.imag)+150)
        
        color = (H,S,V)
        image.putpixel((Px,Py), color)
        image = image.convert("RGB")
image
```

```python
image.save("experiment_4.png")
```
