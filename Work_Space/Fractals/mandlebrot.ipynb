```python
from PIL import Image
from PIL import ImageDraw
```

```python
#Setup the image with appropriate proportions
#Setup drawing area
xa = -2.5
xb = 1
ya = -1.5
yb = 1.5

#setup image size with black background
imgX = 600
imgY = 600
image = Image.new("RGB", (imgX,imgY), (0,0,0))

#set a max iteration
maxIter = 300

#iterate through each pixel
for Py in range(imgY):
    #find the y value in context of the drawing area
    y0 = Py * (yb-ya) / (imgY-1) +ya
    for Px in range(imgX):
        x0 = Px * (xb-xa) / (imgX-1) +xa
        
        c = x0 + y0 * 1j
        
        z = 0
        for i in range(maxIter):
            if abs(z) > 2: break
            z = z*z + c
            color = (i%255, 3*i%255, 5*i%255)     
        image.putpixel((Px,Py), color)
```
