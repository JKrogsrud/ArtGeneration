```python
import numpy as np
import PIL
import random
from PIL import Image
from PIL import ImageDraw
import pandas as pd
```

```python
image = Image.new("RGB", (500,500))
```

```python
image.size
```

```python
### Generate m random coordinates on a PIL Image. Returns as a list.
def random_points(image, m):
    coordinates = []
    for i in range(m):
        x_coord = random.randint(0,image.size[0])
        y_coord = random.randint(0,image.size[1])
        coordinates.append((x_coord,y_coord))
    return coordinates
```

```python
random_points(image,10)
```

```python
coordinates = random_points(image, 10)
df = pd.DataFrame(np.zeros((len(coordinates),len(coordinates)),dtype = np.int8), columns = coordinates, index = coordinates)
df
```

```python
df.at[coordinates[1],coordinates[2]] = 1
df
```
